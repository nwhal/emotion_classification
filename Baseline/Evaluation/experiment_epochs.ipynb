```python
import perceptron_tfidf
import data_utils as du
import evaluator
```

```python
X_train, y_train, X_test, y_test = du.prep_data('isear-train.csv', 'isear-test.csv')
```

```python
f1_scores = []
for i in range(1, 16):
    model = perceptron_tfidf.Perceptron(X_train, y_train, epoch=i)
    model.train()
    predictions = model.predict(X_test)

    eval = evaluator.Evaluator(predictions, y_test)

    eval.ret_fscore()
    avg = sum(eval.f_score.values())/7
    f1_scores.append(avg)
```

```python
print(f1_scores)
```

```python
import matplotlib.pyplot as plt

x = [i for i in range(1, 16)]
y = f1_scores

plt.plot(x,y)
plt.xlabel("Number of epochs")
plt.ylabel("F1 Score")

plt.show()
```
